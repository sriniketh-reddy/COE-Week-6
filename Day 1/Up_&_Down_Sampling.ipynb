```python
import pandas as pd
from sklearn.utils import resample
```

```python
df=pd.DataFrame({
    'Age':[22,25,27,28,30,35,40,45,50,55,60,65,70],
    'Income':[2000,2500,2700,3200,3500,3800,4000,4200,4300,4500,5000,5500,6000],
    'Class':['High','Low','Low','High','High','Low','High','High','Low','Low','High','High','Low']
})
```

```python
df_high = df[df["Class"]=="High"]
df_low = df[df["Class"]=="Low"]
```

```python
df_high_downsampled = resample(df_high,replace=False,n_samples=len(df_low),random_state=50)
df_high_downsampled
```

```python
df_low_upsampled = resample(df_low,replace=True,n_samples=len(df_high))
df_low_upsampled
```

```python
df_resampled = pd.concat([df_high_downsampled,df_low])
df_resampled
```

```python
df_resampled_1 = pd.concat([df_low_upsampled,df_high])
df_resampled_1
```
